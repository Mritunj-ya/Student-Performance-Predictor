```python
categorical_cols=df.select_dtypes(include=["object","string"]).columns
print(categorical_cols)
```

```python
df_encoded=pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True
)
```

```python
df_encoded.head()
```

```python
df_encoded.shape
```

```python
X=df_encoded.drop(columns=["G3"])
Y=df_encoded["G3"]
```

```python
print(X.shape)
print(Y.shape)
print(X.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred[:10])
print(Y_test[:10])
```

```python
from sklearn.metrics import(
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
mae=mean_absolute_error(Y_test,y_pred)
mse=mean_squared_error(Y_test,y_pred)
rmse=mse**0.5
r2=r2_score(Y_test,y_pred)
print("MAE",mae)
print("RMSE",rmse)
print("R2",r2)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
#Create model
tree_model=DecisionTreeRegressor(random_state=42)
#Train
tree_model.fit(X_train,Y_train)
#Predict
tree_pred=tree_model.predict(X_test)
#Evaluate
mae=mean_absolute_error(Y_test,tree_pred)
rmse=mean_squared_error(Y_test, tree_pred)**0.5
r2=r2_score(Y_test,tree_pred)
print("Decision Tree Results")
print("MAE :", mae)
print("RMSE:", rmse)
print("R²  :", r2)

```

```python
print("Training R² :", tree_model.score(X_train, Y_train))
print("Testing R²  :", tree_model.score(X_test, Y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
for depth in[2,5,10]:

    model=DecisionTreeRegressor(max_depth=depth,random_state=42)
    
    model.fit(X_train,Y_train)
    
    train_r2=model.score(X_train,Y_train)
    test_r2=model.score(X_test,Y_test)
    
    
    print(f"Depth:{depth}")
    print(f"Train R2:{train_r2:3f}")
    print(f"Test R2:{test_r2:3f}")
    print("_"*30)

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
#create model
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
#Train
rf_model.fit(X_train,Y_train)
#Predict
rf_pred=rf_model.predict(X_test)
#Evaluate
mae=mean_absolute_error(Y_test,rf_pred)
rmse=mean_squared_error(Y_test,rf_pred)**0.5
r2=r2_score(Y_test,rf_pred)

print("Random Forest Results")
print("MAE :", mae)
print("RMSE:", rmse)
print("R²  :", r2)

print("\nTraining R2:",rf_model.score(X_train, Y_train))
print("Testing R2:",rf_model.score(X_test,Y_test))
```

HyperParameter Tuning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
for n in [10,50,100,200]:
    model=RandomForestRegressor(
        n_estimators=n,
        random_state=42
    )
    model.fit(X_train,Y_train)
    
    train_r2=model.score(X_train,Y_train)
    test_r2=model.score(X_test,Y_test)
    print(f"Trees:{n}")
    print(f"Train R2:{train_r2:.4f}")
    print(f"Test R2:{test_r2:4f}")
    print("-"*30)
```

Cross_Validation

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
#create model
rf=RandomForestRegressor(random_state=42)
#Define the hyoerparameter to try
param_grid={
    "n_estimators":[10,50,100,200],
    "max_depth":[2,4,6,8]
}
#Create grid search 
grid_search= GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)
#Train grid search
grid_search.fit(X_train,Y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)
print("Best Model:", grid_search.best_estimator_)
```

Verifying the best model

```python
best_model=grid_search.best_estimator_
pred=best_model.predict(X_test)
print("MAE:",mean_absolute_error(Y_test,pred))
print("RMSE:",mean_squared_error(Y_test,pred)**0.5)
print("R2:",r2_score(Y_test,pred))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.ensemble import RandomForestRegressor
categorical_features = [
    "school",
    "sex",
    "address",
    "famsize",
    "Pstatus",
    "Mjob",
    "Fjob",
    "reason",
    "guardian",
    "schoolsup",
    "famsup",
    "paid",
    "activities",
    "nursery",
    "higher",
    "internet",
    "romantic"
]
preprocessor= ColumnTransformer(
    transformers=[
        ("cat",OneHotEncoder(handle_unknown="ignore"),
         categorical_features)
    ],
    remainder="passthrough"
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
pipeline=Pipeline([
    ("preprocessor",preprocessor),
    ("model",RandomForestRegressor(random_state=42))
])
pipeline.fit(X_train,Y_train)
pred=pipeline.predict(X_test)
print(r2_score(Y_test,pred))
```
